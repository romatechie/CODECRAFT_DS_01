Step 1 : Importing the basic packages required

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv('API_SP.POP.TOTL_DS2.csv',  encoding='utf-8')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python

df_final = df.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
select_country = 'Afghanistan'


country_data = df_final[df_final['Country Name'] == select_country]


years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {select_country} Over the Years')
```

```python
plt.hist(population, bins=20, color='green', edgecolor='yellow')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(years, population, linestyle='-', color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {select_country} Over the Years')
plt.grid(True)
plt.tight_layout()
plt.show()
```
